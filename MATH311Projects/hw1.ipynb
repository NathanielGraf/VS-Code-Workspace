```python
#1.2 Exercise 2:
def score_rank(scores):
    sorted_scores = sorted(scores, reverse=True)
    
    rank = {}
    for i, score in enumerate(sorted_scores):
        if score not in rank.keys():
            rank[score] = i + 1

    return [rank[score] for score in scores]

print(score_rank([32,87,75,75,50,32]))
```

```python
#1.3 Exercise 3:
def norm_list(nums):
    sorted_nums = sorted(nums)
    min = sorted_nums[0]
    max = sorted_nums[-1]
    
    return [(x - min) / (max - min) for x in nums]

print(norm_list([2,4,10,6,8,4]))
```

```python
#1.4 Exercise 4:
def is_harshad(n):
    """Determine whether or not a number is a Harshad number."""
    n_digits = list(str(n)) # Convert number to list of its digits
    sum_digits = 0 # Create sum variable
    
    for x in n_digits: # Create loop for all digits in n 
        sum_digits += int(x) # Add the digit to the sum of digits (Fixed int(x) error, can't be a str)
        
    return n % sum_digits == 0 # Check whether the number is divisible by this sum, it will return True or False. 

print(is_harshad(18))
print(is_harshad(21))
print(is_harshad(19))
print(is_harshad(1))
```
