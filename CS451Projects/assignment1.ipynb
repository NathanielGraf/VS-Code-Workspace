```python
#Q9
import pandas as pd

df = pd.DataFrame(columns = ["Movie", "Likes", "Dislikes", 'Rating'])
df["Movie"] = ["Cars", "Toy Story", "Finding Nemo", "The Incredibles", "Up",
                "Rogue One", "Wall-E", "Coco", "Moana", "Frozen"]
df["Likes"] = ["Good characters", "Heartwarming story, groundbreaking animation", 
               "Cool journey", "Gritty for a kids movie, great action", "Intro is moving, good redemption arc",
               "The only good new Star Wars movie", "Awesome setting and message", "Beautiful animation, good music",
               "Fantastic music, good Polynesian portrayal", "Great songs and characters"]
df["Dislikes"] = ["Slow plot", "Animation is dated", "Often slow, weak villain", "Goated", 
                  "Too sad", "Too many new characters", "Minimal dialogue", "Predictable plot", "Linear journey", "Overplayed songs"]    
df["Rating"] = [3.2, 4.1, 3.4, 5.0, 4.7, 4.3, 5.0, 3.9, 4.4, 4.5]

print(df.columns)
df
```

```python
#Q10
import numpy as np

new_df = df.copy()

new_df["Movie"] = new_df["Movie"].str.replace(" ", "")
new_df["Likes"] = new_df["Likes"].str.replace(" ", "")
new_df['Dislikes'] = new_df['Dislikes'].str.lower()
new_df["Rating"] = np.floor(new_df["Rating"])
new_df
```

```python
#Q11
import matplotlib.pyplot as plt

plt.hist(df['Rating'], bins=10)
print(df['Rating'].mean())
print(df['Rating'].std())
```

```python
#Q12
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))
ax1.scatter(df.index, df['Rating'], color='blue')
ax1.set_title('Movie Ratings Scatter')
ax1.set_xlabel('Index')
ax1.set_ylabel('Rating')


ax2.bar(df['Movie'], df['Rating'], color='orange')
ax2.set_title('Movie Ratings Bar')
ax2.set_xlabel('Movie')
ax2.set_ylabel('Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Q13
df['ZRatings'] = (df['Rating'] - df['Rating'].mean()) / df['Rating'].std()
plt.hist(df['ZRatings'], bins=10)

```

```python
#Q14
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))
ax1.hist(df['Rating'], bins=10, color='blue')
ax1.set_title('Movie Rating Histogram')
ax1.set_xlabel('Rating')
ax1.set_ylabel('Frequency')


ax2.hist(df['ZRatings'], bins=10, color='blue')
ax2.set_title('Movie ZRatings Histogram')
ax2.set_xlabel('ZRating')
ax2.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#Q15
np.random.seed(1)
df['RandomRatings'] = np.random.uniform(1, 100, size=10)
plt.hist(df['RandomRatings'], bins=10)
print(df['RandomRatings'].mean())
print(df['RandomRatings'].std())

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))
ax1.scatter(df.index, df['RandomRatings'], color='blue')
ax1.set_title('Random Ratings Scatter')
ax1.set_xlabel('Index')
ax1.set_ylabel('Random Rating')

ax2.bar(df['Movie'], df['RandomRatings'], color='orange')
ax2.set_title('Random Ratings Bar')
ax2.set_xlabel('Movie')
ax2.set_ylabel('Random Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

df['ZRandomRatings'] = (df['RandomRatings'] - df['RandomRatings'].mean()) / df['RandomRatings'].std()
plt.hist(df['ZRandomRatings'], bins=10)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))
ax1.hist(df['RandomRatings'], bins=10, color='blue')
ax1.set_title('Random Ratings Histogram')
ax1.set_xlabel('Random Rating')
ax1.set_ylabel('Frequency')


ax2.hist(df['ZRandomRatings'], bins=10, color='blue')
ax2.set_title('ZRandom Ratings Histogram')
ax2.set_xlabel('ZRandom Rating')
ax2.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```
