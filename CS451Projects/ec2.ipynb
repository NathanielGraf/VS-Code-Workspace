```python
game_board = [[None, 2, 1], [1, 2, None,], [None, None, 1]]

def print_board(board):
    for row in board:
        for cell in row: 
            if cell is None:
                print("|_", end ="")
            elif cell == 1:
                print("|o", end ="")
            else:
                print("|x", end ="")
        print("|")
        
print_board(game_board)
```

```python
def add_matrix(mat1, mat2):
    result = []
    for i in range(len(mat1)):
        row = []
        for j in range(len(mat1[0])):
            row.append(mat1[i][j] + mat2[i][j])
        result.append(row)
    return result

mat1 = [[1, 2], [3, 4]]
mat2 = [[6, 0], [0, 6]]

print(add_matrix(mat1, mat2))

mat1 = [[1, 1, 1, 1], [0, 0, 0, 0]]
mat2 = [[6, 0, 6, 0], [0, 0, 0, 0]]

print(add_matrix(mat1, mat2))
```

```python
def mult_matrix(mat1, mat2):
    result = []
    for i in range(len(mat1)):
        row = []
        for j in range(len(mat1[0])):
            row.append(mat1[i][j] * mat2[i][j])
        result.append(row)
    return result

mat1 = [[1, 2], [3, 4]]
mat2 = [[6, 0], [0, 6]]

print(mult_matrix(mat1, mat2))

mat1 = [[1, 1, 1, 1], [0, 0, 0, 0]]
mat2 = [[6, 0, 6, 0], [0, 0, 0, 0]]

print(mult_matrix(mat1, mat2))
```

```python
def mult_matrix_none(mat1, mat2):
    result = []
    for i in range(len(mat1)):
        row = []
        for j in range(len(mat1[0])):
            if mat1[i][j] is None or mat2[i][j] is None:
                return None
            else:
                row.append(mat1[i][j] * mat2[i][j])
        result.append(row)
    return result

mat1 = [[1, 2], [3, 4]]
mat2 = [[6, 0], [0, 6]]

print(mult_matrix_none(mat1, mat2))

mat1 = [[1, 1, 1, 1], [0, 0, 0, 0]]
mat2 = [[6, 0, 6, 0], [0, 0, 0, 0]]

print(mult_matrix_none(mat1, mat2))
```
